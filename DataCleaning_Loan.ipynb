# Checking Missing Data

```python
df.isna().sum()
```

## Gender

```python
df["Gender"].fillna(method="ffill", inplace=True)
df.isna().sum() 
```

The method ffill replaces the NULL values with the value from the previous row.

## Married

```python
df[df["Married"].isna()]
```

```python
df.loc[104, "Married"] = "Yes"
df.loc[228, "Married"] = "No"
df.loc[435, "Married"] = "No"

df.isna().sum()
```

Here, I have set specific null columns to an specific value.

## Dependents

```python
df["Dependents"].unique().tolist()
```

```python
df.groupby(["Dependents"]).count()
```

```python
df.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
```

```python
df.loc[(df["Dependents"].isna()) & (df["Married"] == 'No'), 'Dependents'] = '0'
```

In order to fill the missing values on hte marital status, in the "Dependents" column with '0' but only for rows where "Married" is equal to 'No'. 

```python
df["Dependents"].fillna(method="ffill", inplace=True)
```

```python
df.isna().sum()
```

## Self-Employed

```python
df["Self_Employed"].unique().tolist()
```

```python
df.groupby(["Self_Employed"]).count()
```

```python
df["Self_Employed"].fillna(method="ffill", inplace=True)
```

## LoanAmount

```python
df.describe()
```

```python
df.median()
```

```python
df.loc[df["LoanAmount"].isna(), "LoanAmount"] = 128
df.isna().sum()
```

```python
df.describe()
```

## Loan_Amount_Term

```python
df.loc[df["Loan_Amount_Term"].isna(), "Loan_Amount_Term"] = 360
df.isna().sum()
```

The missing values have been filled with the median value.

## Credit History

```python
df.loc[df["Credit_History"].isna(), "Credit_History"] = 1
df.isna().sum()
```
